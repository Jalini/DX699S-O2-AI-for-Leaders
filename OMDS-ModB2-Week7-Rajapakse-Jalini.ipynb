# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

To  cause viewers to examine the last three months first using preattentive attributes, here are several effective options:

1. Color
Make the last three months’ line segment a contrasting differnt color (e.g., bright red) while the rest remains a neutral gray or light blue.

2.Position (Graph Zoom / Inset)
Create an inset zoom focusing on the last three months in the same figure to direct viewers’ eyes there first.

3.Size / Line Width
Increase line width for the last three months. Thicker lines attract attention quickly.

4. Animation / Motion
Animate the last three months' line segment being drawn after a pause, causing immediate focus.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example dummy data for 3 categories (Miss, Meet, Exceed) across 10 time points
labels = ['Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2']
miss = [5, 6, 8, 10, 12, 15, 20, 25, 33, 42]
meet = [70, 68, 65, 60, 58, 55, 50, 45, 40, 35]
exceed = [25, 26, 27, 30, 30, 30, 30, 30, 27, 23]

x = np.arange(len(labels))

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(10,6))

# Use a color palette that matches the book more closely
miss_color = '#8B0000'       # dark red / maroon
meet_color = '#C0C0C0'       # light grey / silver
exceed_color = '#808080'     # medium grey

ax.bar(x, miss, label='Miss', color=miss_color)
ax.bar(x, meet, bottom=miss, label='Meet', color=meet_color)
ax.bar(x, exceed, bottom=np.array(miss)+np.array(meet), label='Exceed', color=exceed_color)

# Add data labels for Miss % on top of Miss bars
for i, v in enumerate(miss):
    ax.text(i, v/2, f"{v}%", color='white', fontweight='bold', ha='center', va='center')

# Titles and labels
ax.set_title("Goal attainment over time", fontsize=14, fontweight='bold')
ax.set_ylabel("% of total projects")
ax.set_xticks(x)
ax.set_xticklabels(labels)

# Right-aligned annotation similar to original
ax.text(9.5, 105, "As of Q3 2015, more than 1/3 of projects are missing goals", 
        color=miss_color, ha='right', fontsize=12, fontweight='bold')

# Legend
ax.legend(loc='upper left')

# Style adjustments
ax.set_ylim(0,110)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```
